# Relationship between vote count and popularity

```python
import pandas as pd

import tmdb15k.dataset as dataset
import tmdb15k.util as util

df = dataset.load()
df = util.clean(df)
df = util.remove_non_english(df)
df = util.keep_columns(df, ['vote_count', 'popularity'])
```

```python
import tmdb15k.dataset as dataset

# df = dataset.process_column(df, 'popularity', 'popularity', dataset.standardize_z_score())
# df = dataset.process_column(df, 'vote_count', 'vote_count', dataset.standardize_z_score())

# corr = df['popularity'].corr(df['vote_count'])
# print(f"Correlation between popularity and vote count: {corr}")

import matplotlib.pyplot as plt

df = df[df['vote_count'] > 100]
df = df[df['popularity'] < 100]

# print number of movies
print(f"Number of movies: {len(df)}")

plt.scatter(df['popularity'], df['vote_count'])
plt.xlabel('Popularity')
plt.ylabel('Vote Count')
plt.title('Popularity vs. Vote Count')
plt.show()
```
